```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

# 1. Load the road network for Delhi (drivable roads)
place_name = "Delhi, India"
# Configure OSMnx to use a local cache and retry requests
ox.settings.use_cache = True
ox.settings.timeout = 180  # Increase timeout to handle slow responses

G = ox.graph_from_place(place_name, network_type='drive')

# 2. Plot the full road network
ox.plot_graph(G, bgcolor='white', node_size=0, edge_color='gray')

# 3. Define two locations using latitude and longitude
# Example: India Gate and Red Fort
orig_point = (28.6129, 77.2295)  # India Gate
dest_point = (28.6562, 77.2410)  # Red Fort

# 4. Get nearest network nodes to these points
orig_node = ox.nearest_nodes(G, orig_point[1], orig_point[0])
dest_node = ox.nearest_nodes(G, dest_point[1], dest_point[0])

# 5. Find the shortest path (based on distance)
route = nx.shortest_path(G, orig_node, dest_node, weight='length')

# 6. Plot the route
ox.plot_graph_route(G, route, route_color='red', node_size=0, bgcolor='white')
```
